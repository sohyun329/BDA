### Split, Join, Get
- 문자열.split(구분자) : 문자열을 구분자를 기준으로 나눔
- 구분자.join(문자열 or 리스트) : 문자열 사이에 구분자를 추가함
- get : split으로 쪼개진 데이터들은 list형태로 들어가 있고, 이 리스트를 이용해 원하는 값을 출력할 수 있다.

```python
satisfaction_5 = survey.query('satisfaction == 5')
satisfaction_5['comment'].str.split(' ')
```

```python
satisfaction_5['comment'].str.get(1)
```

```python
satisfaction_5['comment'].str.join('*')
```

### Replace
- 문자열.replace(기존값, 변경할 값, 변경횟수) : 변경횟수 만큼 기존 값을 변경할 값으로 대체함.

```python
satisfaction_5['comment'] = satisfaction_5['comment'].str.replace('좋다','good')
satisfaction_5
```

### upper, lower, isupper islower
- upper( ) : 알파벳 소문자 -> 대문자로 변환
- lower( ) : 알파벳 대문자 -> 소문자로 변환
- isupper( ) : 문자열 전체가 대문자인지 판정
- islower( ) : 문자열 전체가 소문자인지 판정

```python
satisfaction_5['comment'].str.upper()
```

```python
print(satisfaction_5['comment'].str.isupper())
```

```python
satisfaction_5['comment'].str.lower()
```

```python
print(satisfaction_5['comment'].str.islower())
```

### strip, lstrip, rstrip
- 문자열.strip(제거할 값) : 양쪽 끝에서 특정 값을 제거
- 문자열.lstrip(제거할 값) : 왼쪽(left) 끝에서 특정 값을 제거
- 문자열.rstrip(제거할 값) : 오른쪽(right) 끝에서 특정 값을 제거

```python
a = '***abc***'
print(a.strip('*'))
print(a.lstrip('*'))
print(a.rstrip('*'))
```

### pad
- 고정길이 데이터를 이용해야 하는 경우, 왼쪽 또는 오른쪽 또는 중앙에 놓고 패딩

```python
satisfaction_5['comment'].str.len().max()
```

```python
satisfaction_5['comment_pad'] = satisfaction_5['comment'].str.pad(width=39,side='left',fillchar='_') # 지정길이 패딩
satisfaction_5['comment_center'] = satisfaction_5['comment'].str.center(width=39,fillchar='_') # 중앙놓고 지정길이 패딩
satisfaction_5['comment_ljust'] = satisfaction_5['comment'].str.ljust(width=39,fillchar='_') # 왼쪽놓고 지정길이 패딩
satisfaction_5['comment_rjust'] = satisfaction_5['comment'].str.rjust(width=39,fillchar='_') # 오른쪽놓고 지정길이 패딩
satisfaction_5['comment_zfill'] = satisfaction_5['comment'].str.zfill(width=39) # 지정길이 0패딩
satisfaction_5
```

### startswith, endswith, contains, match
- 문자열의 시작 값 또는 종료 값 또는 특정 패턴과 일치하는 경우 찾기

```python
new_satifaction_5 = satisfaction_5[['datetime','comment','satisfaction']]
new_satifaction_5
```

```python
#좌측값이 입력패턴과 일치 여부
new_satifaction_5['comment'].str.startswith(pat='시')
```

```python
#우측값이 입력패턴과 일치 여부
new_satifaction_5['comment'].str.endswith(pat='good')
```

```python
#값 중 패턴포함 여부
new_satifaction_5['comment'].str.contains(pat='good', regex=False)
```

```python
#입력패턴과 일치 여부
new_satifaction_5['comment'].str.match(pat='[가-힣]')
```

```python
new_satifaction_5
```

### extract
- str.extract('[정규표현식'])
- 추출하고 싶은 문자열의 정규표현식을 대입하여 데이터추출

```python
new_satifaction_5['year'] = new_satifaction_5['datetime'].str.extract('(\d\d\d\d)')
new_satifaction_5
```
