### 3X3 그래프

```python
from matplotlib import gridspec # gridspec 호출
grid = gridspec.GridSpec(3, 3)
plt.figure(figsize=(20,20))
plt.subplots_adjust(wspace=0.3, hspace = 0.4)

mpg_features = ['manufacturer', 'model', 'displ', 'year', 'cyl', 'trans', 'cty', 'hwy', 'category']

for idx, feature in enumerate(mpg_features):
  ax = plt.subplot(grid[idx])
  ax.set_xticklabels(ax.get_xticklabels(),rotation=45)

  sns.countplot(x = feature, data = mpg, ax = ax)
```

### 2X2 그래프

```python
grid = gridspec.GridSpec(2, 2)
plt.figure(figsize=(20,20))
plt.subplots_adjust(wspace=0.3, hspace = 0.4)

titanic_features = ['Survived', 'Pclass', 'Sex', 'Embarked']

for idx, feature in enumerate(titanic_features):
  ax = plt.subplot(grid[idx])
  ax.set_xticklabels(ax.get_xticklabels(),rotation=45)

  sns.countplot(x = feature, data = titanic, ax = ax)
```
