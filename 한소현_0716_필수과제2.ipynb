- 데이터 0은 부정, 1은 긍정

```python
# .문자 제거, 영어, 공백 제거
movie_rv['document'] = movie_rv['document'].str.replace(pat=r'[^\w]', repl=r' ', regex=True)
movie_rv['document'] = movie_rv['document'].str.replace('[A-Za-z]',"", regex=True)
movie_rv['document'] = movie_rv['document'].str.strip()
movie_rv
```

- 부정, 긍정 데이터 분리

```python
movie_rv_negative = movie_rv[movie_rv['label']==0]
movie_rv_negative.reset_index(drop=False, inplace=True)
movie_rv_negative
```

```python
movie_rv_positive = movie_rv[movie_rv['label']==1]
movie_rv_positive.reset_index(drop=False, inplace=True)
movie_rv_positive
```

```python
# 데이터가 너무 많은 관계로 100개만 추출
movie_0 = movie_rv_negative.sample(n=100)
movie_1 = movie_rv_positive.sample(n=100)
```

```python
from konlpy.tag import Okt
from konlpy.tag import Twitter
```

```python
# 불용어 제거 & 토큰화(명사)
stopwords = pd.read_csv('/content/drive/MyDrive/DATA/stopword.txt',header=None)
stopwords

okt = Okt()
n_words = []

parts = ['Noun']
for i in range(len(movie_0)) :
  text = movie_0['document'].iloc[i]
  words = okt.pos(text)
  words_arr = []
  for n in words :
    if n == 'EOS' or n == '':continue
    words_tmp=n[0]
    part = n[1]
    if not (part in parts) : continue
    if words_tmp in stopwords : continue
    words_arr.append(words_tmp)
  n_words.extend(words_arr)
print(n_words)
```

```python
okt = Okt()
p_words = []

parts = ['Noun']
for i in range(len(movie_1)) :
  text = movie_1['document'].iloc[i]
  words = okt.pos(text)
  words_arr = []
  for n in words :
    if n == 'EOS' or n == '':continue
    words_tmp=n[0]
    part = n[1]
    if not (part in parts) : continue
    if words_tmp in stopwords : continue
    words_arr.append(words_tmp)
  p_words.extend(words_arr)
print(p_words)
```

```python
n_words_df = pd.DataFrame({'words':n_words, 'count':len(n_words)*[1]})
n_words_df
```

```python
nwf = n_words_df.groupby('words').sum().sort_values(by='count',ascending=False).head(15)
nwf
```

```python
p_words_df = pd.DataFrame({'words':p_words, 'count':len(p_words)*[1]})
p_words_df
```

```python
pwf = p_words_df.groupby('words').sum().sort_values(by='count',ascending=False).head(15)
pwf
```

```python
parts =['Noun']
all_words_df = pd.DataFrame()

for n in range(len(movie_0)):
    text = movie_0['document'].iloc[n]
    words = okt.pos(text)
    words_df = pd.DataFrame()
    for i in words:
        if i =='EOS' or i =='':continue
        word_tmp = i[0]
        part = i[1]
        if not (part in parts):continue
        words_df[word_tmp]=[1] #형태소로 나눈 것들은 다 1로 넣어 달라는 것
    all_words_df = pd.concat([all_words_df, words_df],ignore_index =True)
```

```python
all_words_df
```

```python
all_words_df1 = all_words_df.fillna(0)
```

```python
all_words_df1
```

```python
movie_0 = movie_0.dropna()
```

```python
movie_0['document'].iloc[1]
target_text = all_words_df1.iloc[1]
```

```python
target_text
```

```python
import numpy as np
cos_sim=[]
for i in range(len(all_words_df1)):
    cos_text = all_words_df1.iloc[i]
    cos = np.dot(target_text, cos_text) / (np.linalg.norm(target_text)*np.linalg.norm(cos_text))
    cos_sim.append(cos)

all_words_df1['cos_sim'] = cos_sim
```

```python
all_words_df1
```

```python
all_words_df1.sort_values('cos_sim',ascending=False).head()
```
